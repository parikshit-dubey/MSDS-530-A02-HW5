```python
# Multigraph analysis of California HOV lane dataset
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("HOV.csv")
```

```python
print(df)
```

```python
nodes = []

for idx, row in df.iterrows():
    nodes.append(row["Begin_County"])
    nodes.append(row["End_County"])
```

```python
print(nodes)
```

```python
# Creating edges from county to county 
edges = []
for idx, row in df.iterrows():
    edges.append((row["Begin_County"], row["End_County"]))
```

```python
print(edges)
```

```python
# Creating graph of express lane data
G = nx.MultiGraph()
```

```python
# Inserting nodes in the above created graph
for node in nodes:
    G.add_node(node)
```

```python
# Inserting edges in the above created graph
for edge in edges:
    G.add_edge(*edge)
```

```python
print(G)
```

```python

# Create a layout for the graph
pos = nx.spring_layout(G)

plt.figure(figsize=(35, 18))

# Plot the graph
nx.draw_networkx(G, with_labels=True, node_color='grey', node_size=2000, edge_color='blue')

# Show the plot
plt.show()

plt.savefig('HOV.jpeg')
```
